```python
import os
import nltk
import random
```

# Wczytanie korpusu

```python
corpus_path=os.path.abspath("../data/NKJP_1.2_nltk/")
corpus=nltk.corpus.reader.TaggedCorpusReader(root=corpus_path, fileids=".*")
```

```python
len(corpus.words())
```

#### Milionowy, rzeczywiście

```python
print(corpus.sents())
```

## Odczyt otagowanych słów

```python
corpus.tagged_words()
```

## Redukcja tagów do POS

```python
POS_tagged_words=[(word, tag.split(':')[0]) for word, tag in corpus.tagged_words()]   
POS_tags=[tag for word, tag in POS_tagged_words]
```

## Rozkład tagów

```python
fd=nltk.FreqDist(POS_tags)
fd.most_common()
```

## Rozkład zależny tagów i słów

```python
tag_word_distribution=nltk.ConditionalFreqDist([(tag, word) for word, tag in random.sample(POS_tagged_words, 10000)])
tag_word_distribution
```
