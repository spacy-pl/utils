```python
import json
import os
import re
from matplotlib import pyplot as plt
```

```python
def epoch_num(dirname):
    return int(dirname.split('l')[1])


def get_metrics(model_path):
    dirs = os.listdir(model_path)
    epochs_dirs = [s for s in dirs if re.match("^model[0-9]*$", s)]
    epochs_dirs=sorted(epochs_dirs, key=epoch_num)
    metrics = []
    for ed in epochs_dirs:
        metrics.append(json.load(open(model_path+ed+'/accuracy.json')))

    return metrics
```

```python
model_path='../models/trees-pos_LFG_word2vec/'

ms = get_metrics(model_path)
```

```python
plt.plot([m['tags_acc'] for m in ms])
```

```python
plt.plot([m['las'] for m in ms])
```

```python
plt.plot([m['uas'] for m in ms])
```
